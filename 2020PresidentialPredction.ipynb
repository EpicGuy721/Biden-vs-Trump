```python
import numpy as np

import matplotlib.pyplot as plt

#Twitter's positive and negitive words percent
x1_biden = np.array([-0.2, 0.4, -0.2, -0.2, 0.0])
x2_trump = np.array([0.2, 0.0, 0.4, 0.0, -0.4])

# date names for plotting
dates = ["July26,20", "Aug9,20", "Aug22,20", "Oct18,20", "Oct30,20"]
```

```python
# Define plot space
#fig, ax = plt.subplots(figsize=(10, 6))
plt.figure(figsize=(10,6))

# Define x and y axes
plt.plot(dates, x1_biden,'b',label="Biden")
plt.plot(dates, x2_trump,'r',label="Trump")    
        
plt.xlabel("Date")
plt.ylabel("Twitter's positive/negative word percent")

plt.legend(loc='upper left')
```
