```python
# search tweets for keywords
trump_tweets = api.search(q="Trump", count=number_tweets, show_user=False)
biden_tweets = api.search(q="Biden",count=number_tweets, show_user=False)
```

```python
########### build library for stopwrods, positive words, and negative words

# punctuation from string library
import re
from collections import Counter

with open(cur_dir + 'stopwords.txt') as inputfile:
    for line in inputfile:
        stopwords = filter(None, (line.rstrip() for line in inputfile))

#stop = open(cur_dir + "stopwords.txt").read()
#stopwords = stop.split('\n')

 # positive words from: http://www.unc.edu/~ncaren/haphazard/positive.txt
positive_words = open(cur_dir + 'positive.txt').read()
positive = positive_words.split('\n')

# negative words from: http://www.unc.edu/~ncaren/haphazard/negative.txt
negative_words = open(cur_dir + 'negative.txt').read()
negative = negative_words.split('\n')

########### function for preprocessing
#import nltk
from string import punctuation

def turn_to_words(input_list):
    text = ' '.join(input_list)  # merge list of strings into a single big string, separated by ' '
    
    for p in list(punctuation): # remove punctuation
        text = text.replace(p,' ')
        
    #for p in list(stopwords): # remove stopwords
    #    text = text.replace(p,' ')
    
    text = text.lower() # to lowercase
    text = text.split() # turn into words
    
    return text
```

```python
#for tweet in Trump_tweets:

mylist = []
for tweet in trump_tweets:
    mylist.append(tweet.text) # create a list of tweets (string)

words = turn_to_words(mylist) # create list of words out of all the tweets (remove punctuation + lowercase)

sentiment = [];
for word in words:
    if word in list(positive) or word in list(negative):
        sentiment.append(word)  # create a list of sentiment words

#print(sentiment)

sentiment_counts = Counter(sentiment)
ten_common = sentiment_counts.most_common(10) # 10 most common sentiment words
                                                                                                                

ten_common_list = []
for i in ten_common:
    ten_common_list.append(i[0]) # extract words from each tuple, and make them into a list

pos = 0
neg = 0
for word in ten_common_list:
    if word in positive:
        pos = pos + 1
    if word in negative:
        neg = neg + 1
print(ten_common)
print ("TRUMP: positive words: " + str(pos) + " negative words: " + str(neg))
x=(pos-neg)/10.0
print ("TRUMP's Average Sentiment Value: " + str(float("{0:.2f}".format(x))))    
```
